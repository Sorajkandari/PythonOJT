```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    return re.sub(CLEANR, '', text)
```

```python
df2['PC_ITEM_DESC_SMALL'] = df2['PC_ITEM_DESC_SMALL'].apply(remove_html_tags)
```
