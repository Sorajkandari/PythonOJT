```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['PC_ITEM_DESC_SMALL'] = df['PC_ITEM_DESC_SMALL'].apply(remove_html_tags)
df2 = df
```
